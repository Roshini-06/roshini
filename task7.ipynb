```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting

# To Load the Pima Indians Diabetes dataset
diabetes_data = pd.read_csv('/content/drive/MyDrive/diabetes (1).csv')

# To Explore the dataset
print(diabetes_data.head())



```

i)CORRELATION AND SCATTER PLOTS

```python
#  Correlation and scatter plots
correlation_matrix = diabetes_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatter plot for two variables
sns.scatterplot(x='Glucose', y='Insulin', data=diabetes_data)
plt.title('Scatter Plot of Glucose vs Insulin')
plt.show()
```

ii)HISTOGRAM

```python
#  Histograms
diabetes_data.hist(figsize=(12, 10))
plt.suptitle('Histograms of Pima Indians Diabetes Data')
plt.show()
```

iii)THREE DIMENSIONAL PLOTTING

```python

#  Three-dimensional plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(diabetes_data['Glucose'], diabetes_data['Insulin'], diabetes_data['BMI'], c='r', marker='o')
ax.set_xlabel('Glucose')
ax.set_ylabel('Insulin')
ax.set_zlabel('BMI')
plt.title('3D Scatter Plot')
plt.show()
```

d. Write a NumPy program to find the real and imaginary parts of an array of complex
numbers

```python
import numpy as np

# Create an array of complex numbers
complex_array = np.array([1 + 2j, 3 - 4j, 5 + 6j])

# Find the real and imaginary parts
real_parts = np.real(complex_array)
imaginary_parts = np.imag(complex_array)

# Print the results
print("Complex Array:", complex_array)
print("Real Parts:", real_parts)
print("Imaginary Parts:", imaginary_parts)
```
